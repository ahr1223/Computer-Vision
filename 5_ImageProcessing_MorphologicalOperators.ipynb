```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
image1 = cv2.imread('text.png',0)
plt.imshow(image1,cmap='gray')
```

```python
kernel = np.ones((5,5),dtype=np.uint8)

result = cv2.erode(image1,kernel,iterations=1)
plt.imshow(result,cmap='gray')
```

```python
kernel = np.ones((5,5),dtype=np.uint8)

result = cv2.erode(image1,kernel,iterations=4)
plt.imshow(result,cmap='gray')
```

```python
kernel = np.ones((5,5),dtype=np.uint8)

result = cv2.dilate(image1,kernel,iterations=1)
plt.imshow(result,cmap='gray')
```

```python
white_noise = np.random.randint(low=0,high=2,size=(600,600))
white_noise = white_noise*255
plt.imshow(white_noise,cmap='gray')
```

```python
image1
```

```python
white_noise
```

```python
noise_img = white_noise+image1
plt.imshow(noise_img,cmap='gray')
```

```python
noise_img
```

```python
noise_img=noise_img.astype(np.uint8)
opening = cv2.morphologyEx(noise_img,cv2.MORPH_OPEN,kernel)
plt.imshow(opening,cmap='gray')
```

```python
black_noise = np.random.randint(low=0,high=2,size=(600,600))*255
black_noise_img = image1 + black_noise
black_noise_img[black_noise_img==255]=0 
```

```python
cv2.imwrite('MorphImage1.jpeg',black_noise_img)
```

```python
plt.imshow(black_noise_img,cmap='gray')
```

```python
black_noise_img=black_noise_img.astype(np.uint8)
closing = cv2.morphologyEx(black_noise_img,cv2.MORPH_CLOSE,kernel)
plt.imshow(closing,cmap='gray')
```

```python
gradient = cv2.morphologyEx(image1,cv2.MORPH_GRADIENT,kernel)
plt.imshow(gradient,cmap='gray')
```
