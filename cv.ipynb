```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv2.imread("/content/levi-ackerman-3840x2160-10446.jpg")
plt.imshow(img)
```

```python
# bgr to rgb
img_rgb = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
plt.axis("off")
plt.show()
```

```python
# different component rgb
r,g,b = cv2.split(img_rgb)
plt.imshow(r,cmap="Reds")
plt.axis("off")
plt.show()
```

```python
plt.imshow(g,cmap="Greens")
plt.axis("off")
plt.show()
```

```python
plt.imshow(b,cmap="Blues")
plt.axis("off")
plt.show()
```

```python
#convert img to gray
img_gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
plt.imshow(img_gray,cmap="gray")
plt.axis("off")
plt.show()
```

```python
#resize img
re = cv2.resize(img_rgb,(150,300))
plt.imshow(re)
plt.axis("off")
plt.show()
```

```python
#img translate
row,col = img.shape[:2]
M = np.float32([[1,0,1000],[0,1,500]])
tras = cv2.warpAffine(img,M,(col,row))
plt.imshow(tras)
plt.axis("off")
plt.show()
```

```python
#img rotate
center = (row//2,col//2)
M = cv2.getRotationMatrix2D(center,45,1)
ro = cv2.warpAffine(img_rgb,M,(col,row))
plt.imshow(ro)
plt.axis("off")
plt.show()
```

```python
#negative img
neg = 255-img_rgb
plt.imshow(neg)
plt.axis("off")
plt.show()
```

```python
#log trans
img_gray = img_gray.astype(np.float32)
c = 255/np.log(1+np.max(img_gray))
img_log = c*(np.log(img_gray+1))
img_log = np.array(img_log,dtype=np.uint8)
plt.imshow(img_log,cmap="gray")
plt.axis("off")
plt.show()
```

```python
#gama trans
gama  = 2.0
img_gama = np.array(255*(img_gray/255)**gama,dtype="uint8")
plt.imshow(img_gama,cmap="gray")
plt.axis("off")
plt.show()
```

```python
#hist equalizer
img_gray = img_gray.astype(np.uint8)
e = cv2.equalizeHist(img_gray)
plt.imshow(e,cmap="gray")
plt.axis("off")
plt.show()
```

```python
#morphological
kernel = np.ones((5,5),np.uint8)
ero = cv2.erode(img_gray,kernel,1)
dul = cv2.dilate(img_gray,kernel,1)
open = cv2.morphologyEx(img_gray,cv2.MORPH_OPEN,kernel)
close = cv2.morphologyEx(img_gray,cv2.MORPH_CLOSE,kernel)
plt.imshow(ero,cmap="gray")
plt.axis("off")
plt.show()
```

```python
plt.imshow(dul,cmap="gray")
plt.axis("off")
plt.show()
```

```python
plt.imshow(open,cmap="gray")
plt.axis("off")
plt.show()
```

```python
plt.imshow(close,cmap="gray")
plt.axis("off")
plt.show()
```
